### Loading in image

```python
# Call libraries needed for visualizing images
import skimage.io as io
import skimage.filters
import numpy as np # this allows for a library to be "nicknamed" for easy calling in code
from matplotlib import pyplot as plt
from IPython.display import Image
# Not all coding languages require this type of library setup but for python this is the case.
```

```python
filename='525_neurons.png'#save variable with name of image; ' ' used to denote a string (aka text)
img =  io.imread(filename) #reads in image using a function called "imread" that is called from the io library
plt.imshow(img) #Visualize image using a plot from matplotlib library
```

#### Now that we have an image, what can we look for?
##### We can count cells, we can identify different cells based on color, we can find average sizing of the cells etc.
#### What do you think the image is showing?
##### Red is axon skeleton, blue is nuclear data, green is myelin sheaths (found on some neurons).
#### Is this normal tissue?
##### Actually no, this is damaged tissue hence the structural elements look different (circular and nuclear data is spread out along with the axon skeleton too).


```python
# What other information we can get out of the images?
print('Shape of the image',img.shape)
print('Data type of image',img.dtype)
#This is useful because depending on your data type and image size, you will manipulate the data differently.
# Also, you may run into errors if certain functions cannot process certain types and this is helpful to know prior.
```

### Converting image to array

```python
img_ary = np.asarray(img) # Creating an array of the image
#print('Array of Image',img_ary)
#print('Raw Image', img)
```

### Why do we need to convert image to an array?
#### This is just a display of the matrix, we want to be able to manipulate it with varying functions. This can only be done if we convert to an array!

### RGB Values 
#### Pixel Values and their Locations

```python
# What color do we have in the image?
unique_pixel_values = np.unique(img_ary)
print(unique_pixel_values)
# Notice this is a 256 pixel size (RGB).
```

# Color chart
https://www.rapidtables.com/web/color/RGB_Color.html #Modify this to a markdown hyperlink
# Add a plot box that they can change the color values in

```python
r = img[:,:,0]
g = img[:,:,1]
b = img[:,:,2]
plot1 = plt.figure(1)
plt.imshow(r, cmap='Reds')
plot2 = plt.figure(2)
plt.imshow(b, cmap='Blues')
plot3 =plt.figure(3)
plt.imshow(g, cmap='Greens')
```

```python
#As you can see, python can become very complex and you can create special for loops to iterate this process
fig, ax = plt.subplots(1, 4, figsize = (18, 30))
ax[0].imshow(img_ary) 
ax[0].axis('off')
ax[0].set_title('original RGB')
#for loop allows you to iterate multiple times to repeat a command
for i, cmap in enumerate(['Reds','Greens','Blues']):
    ax[i+1].imshow(img_ary[:,:,i], cmap=cmap) 
    ax[i+1].axis('off')
    ax[i+1].set_title(cmap[0])
plt.show()
```
