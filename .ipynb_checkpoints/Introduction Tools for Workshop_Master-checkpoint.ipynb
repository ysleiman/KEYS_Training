## Python programs are plain text files.

*   They have the `.py` extension to let everyone (including the operating system) 
    know it is a Python program.
    *   This is convention, not a requirement.
*   It's common to write them using a text editor but we are going to use the [Jupyter Notebook](http://jupyter.org/)
*   The bit of extra setup is well worth it because the Notebook provides code completion 
    and other helpful features.
*   Notebook files have the extension `.ipynb` to distinguish them from plain-text Python programs.
    * *These files types along with .pngs or .jpgs will be the majority of the file types you will be saving to your github for today's session.*

# Advantages of Using Jupyter Notebook to run Python   
* You can easily type, edit, and copy and paste blocks of code.
* Tab completion allows you to easily access the names of things you are 
    using and learn more about them.
	- It allows you to annotate your code with links, different sized text, bullets, 
    etc to make it more accessible to you and your collaborators.
	- It allows you to display figures next to the code that produces them to 
    tell a complete story of the analysis.
*   The notebook is stored as JSON but can be saved as a .py file if you would
    like to run it from the bash shell or a python interpreter.
*   Just like a webpage, the saved notebook looks different to what you see when 
    it gets rendered by your browser.
    *   But this format allows Jupyter to mix software (in several languages) with documentation 
      and graphics, all in one file.

### Loading in image

```python
# Call libraries needed for visualizing images
import skimage.io as io
import skimage.filters
import numpy as np # this allows for a library to be "nicknamed" for easy calling in code
from matplotlib import pyplot as plt
from IPython.display import Image
# Not all coding languages require this type of library setup but for python this is the case.
```

```python
filename='525_neurons.png'#save variable with name of image; ' ' used to denote a string (aka text)
img =  io.imread(filename) #reads in image using a function called "imread" that is called from the io library
plt.imshow(img) #Visualize image using a plot from matplotlib library
```

#### Now that we have an image, what can we look for?
##### We can count cells, we can identify different cells based on color, we can find average sizing of the cells etc.
#### What do you think the image is showing?
##### Red is axon skeleton, blue is nuclear data, green is myelin sheaths (found on some neurons).
#### Is this normal tissue?
##### Actually no, this is damaged tissue hence the structural elements look different (circular and nuclear data is spread out along with the axon skeleton too).


```python
# What other information we can get out of the images?
print('Shape of the image',img.shape)
print('Data type of image',img.dtype)
#This is useful because depending on your data type and image size, you will manipulate the data differently.
# Also, you may run into errors if certain functions cannot process certain types and this is helpful to know prior.
```

### Converting image to array

```python
img_ary = np.asarray(img) # Creating an array of the image
#print('Array of Image',img_ary)
#print('Raw Image', img)
```

### Why do we need to convert image to an array?
#### Array objects let you do important operations such as matrix multiplication, transposition,solving equation systems, vector multiplacation, and normalization. 
#### This allows for aligning and warping of images. 
#### This is just a display of the matrix, we want to be able to manipulate it with varying functions. This can only be done if we convert to an array!

```python
r = img[:,:,0]
g = img[:,:,1]
b = img[:,:,2]
plot1 = plt.figure(1)
plt.imshow(r, cmap='Reds')
plot2 = plt.figure(2)
plt.imshow(b, cmap='Blues')
plot3 =plt.figure(3)
plt.imshow(g, cmap='Greens')
```

```python
#As you can see, python can become very complex and you can create special for loops to iterate this process
fig, ax = plt.subplots(1, 4, figsize = (18, 30))
ax[0].imshow(img_ary) 
ax[0].axis('off')
ax[0].set_title('original RGB')
#for loop allows you to iterate multiple times to repeat a command
for i, cmap in enumerate(['Reds','Greens','Blues']):
    ax[i+1].imshow(img_ary[:,:,i], cmap=cmap) 
    ax[i+1].axis('off')
    ax[i+1].set_title(cmap[0])
plt.show()
```

# Binarizing Image & Histograms

## What is an binary image? What are you doing when you binarize an image?

### Ans: A binary image is an image that consists of only two colors (commonly white and black).
### Ans: When you binarize the image you apply an algorith that will will convert each pixel to a value of either 1 or 0.

```python
#Binary Thresholding
#Understanding Histograms

import matplotlib.pyplot as plt
from skimage import data
from skimage.filters import threshold_mean #import selected thresholding method

thresh = threshold_mean(r) #establish and register thresholding method with image  
binary = r > thresh #applys threshold to image to create binary image.

#Code for plots and shapping
fig, axes = plt.subplots(ncols=3, figsize=(12, 2.5))
ax = axes.ravel()

ax[0] = plt.subplot(1, 3, 1)
ax[1] = plt.subplot(1, 3, 2)
ax[2] = plt.subplot(1, 3, 3, sharex=ax[0], sharey=ax[0])

#Displaying first image
ax[0].imshow(r, cmap=plt.cm.gray)
ax[0].set_title('Original')
ax[0].axis('off')

#Creating histogram
ax[1].hist(r.ravel(), bins=256)
ax[1].set_title('Histogram')
ax[1].axvline(thresh, color='r')

#Displaying binary image
ax[2].imshow(binary, cmap=plt.cm.gray)
ax[2].set_title('Thresholded')
ax[2].axis('off')

plt.show()
```

```python
#Display all thresholding methods in skiimage library package
from skimage.filters import try_all_threshold
fig, ax = try_all_threshold(r, figsize=(10, 10), verbose=False)
plt.show()
```
